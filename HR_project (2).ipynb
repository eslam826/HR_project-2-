```python

#"Replace the satisfied numbers in the columns for EnvironmentSatisfaction, JobSatisfaction, and RelationshipSatisfaction with their corresponding values."

# Create a mapping dictionary for the satisfaction ratings
satisfaction_mapping = {
    1: 'Very Dissatisfied',
    2: 'Dissatisfied',
    3: 'Neutral',
    4: 'Satisfied',
    5: 'Very Satisfied',
}

# Apply the mapping to the columns
performance_rating['EnvironmentSatisfaction'] = performance_rating['EnvironmentSatisfaction'].map(satisfaction_mapping)
performance_rating['JobSatisfaction'] = performance_rating['JobSatisfaction'].map(satisfaction_mapping)
performance_rating['RelationshipSatisfaction'] = performance_rating['RelationshipSatisfaction'].map(satisfaction_mapping)


performance_rating
```

```python
#repalce rating numbers by value
Rating_level_mapping = {
    1: 'Unacceptable',
    2: 'Needs Improvement',
    3: 'Meets Expectation',
    4: 'Exceeds Expectation',
    5: 'Above and Beyond',
}

# Apply the mapping to the columns
performance_rating['SelfRating'] = performance_rating['SelfRating'].map(Rating_level_mapping)
performance_rating['ManagerRating']= performance_rating['ManagerRating'].map(Rating_level_mapping)
performance_rating

```

```python
#repalce rating numbers by value

EducationLevel_mapping ={
    1: 'No Formal Qualifications',
    2: 'High School',
    3:	'Bachelors',
    4: 'Masters',
    5:	'Doctorat'

}

# Apply the mapping to the columns
Employee['Education'] = Employee['Education'].map(EducationLevel_mapping)

Employee
```

**1..look at employees satisfaction and performance rating percentage**

```python
#satisfied and rating columns'percentage
EnvironmentSatisfaction=(all_employee_data['EnvironmentSatisfaction'].value_counts(normalize=True) * 100).round(2)
JobSatisfaction=(all_employee_data['JobSatisfaction'].value_counts(normalize=True) * 100).round(2)
RelationshipSatisfaction=(all_employee_data['RelationshipSatisfaction'].value_counts(normalize=True) * 100).round(2)
SelfRating=(all_employee_data['SelfRating'].value_counts(normalize=True) * 100).round(2)
ManagerRating=(all_employee_data['ManagerRating'].value_counts(normalize=True) * 100).round(2)

print(EnvironmentSatisfaction)
print('\n')
print(JobSatisfaction)
print('\n')
print(RelationshipSatisfaction)
print('\n')
print(SelfRating)
print('\n')
print(ManagerRating)
```

```python
fig,axes=plt.subplots(3,2 ,figsize=(25,10 ))
fig.patch.set_facecolor('#f2f2f2')

#Environment Satisfaction
axes[0,0].bar(EnvironmentSatisfaction.index , EnvironmentSatisfaction.values , color=cust_color)
axes[0,0].set_ylabel('percenrag%' ,weight='semibold',fontname='monospace',size=10)
axes[0,0].set_xlabel('Environment Satisfaction',weight='semibold',fontname='monospace',size=10)
axes[0,0].spines['left'].set_color('black')
axes[0,0].spines['top'].set_visible(False)
axes[0,0].spines['right'].set_visible(False)
#Job Satisfaction
axes[1,0].bar(JobSatisfaction.index , JobSatisfaction.values , color=cust_color)
axes[1,0].set_ylabel('percenrag%' ,weight='semibold',fontname='monospace',size=10)
axes[1,0].set_xlabel('Job Satisfactionn',weight='semibold',fontname='monospace',size=10)
axes[1,0].spines['bottom'].set_color('black')
axes[1,0].spines['left'].set_color('black')
axes[1,0].spines['top'].set_visible(False)
axes[1,0].spines['right'].set_visible(False)

#RelationshipSatisfaction
axes[0,1].bar(RelationshipSatisfaction.index , RelationshipSatisfaction.values , color=cust_color)
axes[0,1].set_ylabel('percenrag%' ,weight='semibold',fontname='monospace',size=10)
axes[0,1].set_xlabel('RelationshipSatisfaction',weight='semibold',fontname='monospace',size=10)
axes[0,1].spines['bottom'].set_color('black')
axes[0,1].spines['left'].set_color('black')
axes[0,1].spines['top'].set_visible(False)
axes[0,1].spines['right'].set_visible(False)

#SelfRating
axes[1,1].bar(SelfRating.index , SelfRating.values , color=cust_color)
axes[1,1].set_ylabel('percenrag%' ,weight='semibold',fontname='monospace',size=10)
axes[1,1].set_xlabel('SelfRating',weight='semibold',fontname='monospace',size=10)
axes[1,1].spines['bottom'].set_color('black')
axes[1,1].spines['left'].set_color('black')
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['right'].set_visible(False)

# ManagerRating
axes[2, 0].bar(ManagerRating.index, ManagerRating.values, color=cust_color)
axes[2, 0].set_ylabel('percenrag%', weight='semibold', fontname='monospace', size=10)
axes[2, 0].set_xlabel('ManagerRating', weight='semibold', fontname='monospace', size=10)
axes[2, 0].spines['bottom'].set_color('black')
axes[2, 0].spines['left'].set_color('black')
axes[2, 0].spines['top'].set_visible(False)
axes[2, 0].spines['right'].set_visible(False)

# إزالة المحور الفارغ
axes[2, 1].axis('off')
plt.suptitle('Employee satisfaction and Performance Rating',weight='bold',fontname='monospace',size=30)

plt.show()
```

**3-attrion_rat percentag**

```python
#attrion_rat percentag

attrition_rate = (all_employee_data['Attrition'].value_counts(normalize=True) * 100).round(2)
print(attrition_rate)

# Create a pie chart
plt.figure(figsize=(5, 8))
plt.style.use('ggplot')

labels=['stay','leave']
colors=['green','red']
explode=[0,0.2]

plt.pie(attrition_rate,labels=labels ,colors=colors,explode=explode,shadow = True ,startangle=90 ,autopct='%1.2f%%', wedgeprops={'edgecolor':'black'})

plt.title('attrition_rate')
plt.show()
```

**4..correlation between attrition_job_satisfaction **



```python
# إنشاء جدول تقاطعي لحساب نسبة الدوران الوظيفي لكل مستوى من مستويات الرضا الوظيفي
attrition_job_satisfaction = pd.crosstab(all_employee_data['JobSatisfaction'], all_employee_data['Attrition'], normalize='index') * 100
attrition_job_satisfaction = attrition_job_satisfaction.rename(columns={'Yes': ' leaved', 'No': 'stayed'})
print(attrition_job_satisfaction)

# إنشاء مخطط شريطي مزدوج للعلاقة بين الرضا الوظيفي والدوران الوظيفي
attrition_job_satisfaction.plot(kind='bar', figsize=(10,7), stacked=False, color=['red', 'green'])
plt.title('correlation between attrition_job_satisfaction ')
plt.xlabel('job_satisfaction')
plt.ylabel('percentage %')
plt.legend(title='Attrition')
plt.show()

```


**5..self_rating and manager_rating_distribution**



```python
#self_rating and manager_rating_distribution

self_rating_distribution = all_employee_data['SelfRating'].value_counts(normalize=True) * 100
manager_rating_distribution = all_employee_data['ManagerRating'].value_counts(normalize=True) * 100

print("توزيع التقييم الذاتي:")
print(self_rating_distribution)

print("\nتوزيع تقييم المديرين:")
print(manager_rating_distribution)


fig,axes=plt.subplots(1,2 ,figsize=(25,10 ))
fig.patch.set_facecolor('#f2f2f2')

# self_rating_distribution
axes[0].bar(x=self_rating_distribution.index, height=self_rating_distribution.values, color=cust_color)  # Changed y to height
axes[0].set_ylabel('percenrag%', weight='semibold', fontname='monospace', size=10)
axes[0].set_xlabel('self_rating_distribution Satisfaction', weight='semibold', fontname='monospace', size=10)
axes[0].spines['left'].set_color('black')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

# manager_rating_distribution
axes[1].bar(x=manager_rating_distribution.index, height=manager_rating_distribution.values, color=cust_color)  # Changed y to height
axes[1].set_ylabel('percenrag%', weight='semibold', fontname='monospace', size=10)
axes[1].set_xlabel('manager_rating_distribution', weight='semibold', fontname='monospace', size=10)
axes[1].spines['bottom'].set_color('black')
axes[1].spines['left'].set_color('black')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
```

```python
# create a heatmap to show corr between manger and self rating
# إنشاء جدول تقاطعي
rating_crosstab = pd.crosstab(all_employee_data['SelfRating'], all_employee_data['ManagerRating'], normalize='index') * 100

# (Heatmap)
plt.figure(figsize=(10,8))
sns.heatmap(rating_crosstab, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('corr between manger and self rating(%)')
plt.xlabel('manager rating')
plt.ylabel(' self rayig')
plt.show()
```

**6_Average Salary vs. Years at Company**

```python
#years at company ditribution

plt.figure(figsize=(10,6))
sns.histplot(all_employee_data['YearsAtCompany'], bins=10, kde=True, color='blue')
plt.title(' YearsAtCompany dis ')
plt.xlabel(' YearsAtCompany')
plt.ylabel('count ')
plt.show()
```

```python
#Average Salary vs. Years at Company

# Calculate average salary for each year at the company
average_salary_by_year = all_employee_data.groupby('YearsAtCompany')['Salary'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='YearsAtCompany', y='Salary', data=average_salary_by_year)
plt.title("Average Salary vs. Years at Company")
plt.show()
```

**7_Average Salary vs. jop role**

```python
#Average Salary vs. jop role
average_salary_by_year = all_employee_data.groupby('JobRole')['Salary'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
# Use matplotlib.pyplot.pie instead of seaborn.pie
plt.pie(average_salary_by_year['Salary'], labels=average_salary_by_year['JobRole'], autopct='%1.1f%%')
plt.title("Average Salary vs. jop rol")
plt.show()
```

**8_Average Salary vs. jopsatisfiction**

```python
#Average Salary vs. jopsatisfiction
average_salary_by_year = all_employee_data.groupby('JobSatisfaction')['Salary'].mean().reset_index()

# Create the plot
plt.plot(average_salary_by_year['JobSatisfaction'], average_salary_by_year['Salary'], marker='o')
plt.title("Average Salary vs. JobSatisfaction")
plt.xlabel("JobSatisfaction")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

**9_ages distribution through gender**


```python
#ages distribution through gender

average_salary_by_year = all_employee_data.groupby('Gender')['Age'].mean().reset_index()

# Create the bar plot
sns.barplot(x='Gender', y='Age', data=average_salary_by_year)
plt.title("ages distribution through gender")
plt.show()
```
